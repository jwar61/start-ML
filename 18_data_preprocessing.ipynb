```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data

X_minmax = MinMaxScaler().fit_transform(x) #X minus Min dived by Max minus Min
X_stand = StandardScaler().fit_transform(x) #X minus avarage dived by equar type
X_Robust= RobustScaler().fit_transform(x)#X minus Mediane dived by IQR

plt.scatter(x[:,2],x[:,3], label='original')
plt.scatter(X_minmax[:,2],X_minmax[:,3],label='MiniMax')
plt.scatter(X_stand[:,2],X_stand[:,3],label='Standardist')
plt.scatter(X_Robust[:,2],X_Robust[:,3],label='Robust')
plt.legend()
```

```python
#Feature Engineering
import numpy as np
from sklearn.linear_model import LinearRegression

x= np.linspace(0, 4, 100).reshape((100,1))
y= x**2 + 5*np.cos(x) + np.random.randn(100,1)

plt.scatter(x,y)

model = LinearRegression().fit(x,y)
y_pred = model.predict(x)

plt.plot(x,y_pred, c='r',lw=3)
```

```python
from sklearn.preprocessing import PolynomialFeatures
plt.scatter(x,y)

x_poly = PolynomialFeatures(3).fit_transform(x) #create an foction polynomial

model = LinearRegression().fit(x_poly ,y)
y_pred = model.predict(x_poly )

plt.plot(x,y_pred, c='r',lw=3)
```
