```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, centers=3, random_state=42)
plt.scatter(X[:,0],X[:,1])    
```

```python
#CLUSTERING
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=model.fit_predict(X))
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], 
            marker='x', s=200, linewidths=3, color='red')
model.inertia_ #calculate the distace between the point within the cluster and the point center
```

```python
#Elbow Method
inertia = []
k_range = range(1,20)
for k in k_range:
    model = KMeans(n_clusters = k).fit(X)
    inertia.append(model.inertia_)
    
```

```python
plt.plot(k_range, inertia)
plt.xlabel('n cluster')
plt.ylabel('cout du model (inertia)')
```

```python
import numpy as np
x, y = make_blobs(n_samples=100, centers=1, random_state=42 , cluster_std=0.5)
x[-1,:]= np.array([2.25,5])
```

```python
plt.scatter(x[:,0],x[:,1]) 
```

```python
from sklearn.ensemble import IsolationForest
```

```python
model = IsolationForest(contamination=0.01)
model.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1], c=model.predict(x)) 
```
